# 6. Read BM_lowCont image and plot the histogram and perform Equalization

````python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
````

````python
#Import image 
BM=cv2.imread("BM_lowCont.png")
````

````python
#Split the image in RGB channel
b,g,r= cv2.split(BM)
````

````python
#Calculate histograms for each color channel
hist_b = cv2.calcHist([b],[0],None,[256],[0,256])
hist_g= cv2.calcHist([g],[0],None,[256],[0,256])
hist_r= cv2.calcHist([r],[0],None,[256],[0,256])
````

````python
#plot histogram for each channel
plt.figure(figsize=(15,8))
plt.subplot(131), plt.plot(hist_b, color='blue')
plt.title("Blue Channel")
plt.subplot(132),plt.plot(hist_g,color="green")
plt.title("Green Channel")
plt.subplot(133),plt.plot(hist_r,color="red")
plt.title("Red Channel")
plt.show()
````

````python
#converting the image to greyscale
gray = cv2.cvtColor(BM, cv2.COLOR_BGR2GRAY)
````

````python

#performing equalization
eq = cv2.equalizeHist(gray)
````

````python
#displaying the original and Equalized image together
fig, ax=plt.subplots(1,2,figsize=(15,5))
ax[0].imshow(cv2.cvtColor(BM, cv2.COLOR_BGR2RGB))
ax[0].set_title("ORIGINAL")
ax[1].imshow(cv2.cvtColor(eq, cv2.COLOR_BGR2RGB))
ax[1].set_title("Equalized")
````

# 7. Read bridge image and add salt & pepper noise

````python
#importing important libraries
import cv2
import random
````

````python
#impoting the bridge image and reading it
bridge=cv2.imread("bridge.png")
````

````python
#converting the image to greyscale
gray=cv2.cvtColor(bridge,cv2.COLOR_BGR2GRAY)
````

````python
#checking the shape of image
gray.shape
````

````python
##ADDING salt and pepper noise
#Defining a fucntion that takes an gray scale image and convert it into noisey image by adding salt and pepper noise
#Random number of pixels to the maximum value (255) and another random number of pixels to the minimum value (0
def noisey(img):
    #Getting dimension of image
    row, col = img.shape
    #Generating random num of pixel to add for white noise 
    no_of_pixels_white = random.randint(200, 8000)
    #Loop to pick random pixel from image shape and make it white
    for i in range(no_of_pixels_white):
        y_axis = random.randint(0, row - 1)
        x_axis = random.randint(0, col - 1)
        img[y_axis][x_axis] = 255
    #Generating random num of pixel to add for black noise
    no_of_pixels_black = random.randint(200, 8000)
    #Loop to pick random pixel from image shape and make it black
    for i in range(no_of_pixels_black):
        y_axis_b = random.randint(0, row - 1)
        x_axis_b = random.randint(0, col - 1)
        img[y_axis_b][x_axis_b] = 0
    return img
````

````python
#calling the function with our converted gray scale image
noisey(gray)
````

````python
#displaying the noisey image using matplotlib
plt.imshow(cv2.cvtColor(noisey(gray), cv2.COLOR_BGR2RGB))
````

# 9. Add random and Gaussian noise

````python
import cv2
import random
import numpy as np
````

````python
#importing the original image
bridge=cv2.imread("bridge.png")
#cv2.imshow('bridge image',bridge)
#cv2.waitKey()
#cv2.destroyAllWindows()
````

````python
#converting the image to grey scale
gray=cv2.cvtColor(bridge,cv2.COLOR_BGR2GRAY)
````

### Adding Random noise


````python
#Defining a function that take input image and generate random noise as the shape of the image.
#The data type is set to np.uint8 and both original image and noise is added later
def add_rand_noise(image):
    #getting  row and column from image
    row,col=image.shape
    #generating noise using numpy under 0-255 and under the size of image
    noise=np.random.randint(0,256,(row,col),dtype=np.uint8)
    #adding the noise and original image
    image=cv2.add(image,noise)
    return image
````

````python
add_rand_noise(gray)
````

````python
#displaying the combined image of noise and original image using opencv
plt.imshow(cv2.cvtColor(add_rand_noise(gray), cv2.COLOR_BGR2RGB))
````

###### Adding gaussain noise in greyscale to an image

````python
import cv2
import numpy as np

img = cv2.imread('bridge.png')
#Generate Gaussian noise
gauss = np.random.normal(0,1,img.size)
gauss = gauss.reshape(img.shape[0],img.shape[1],img.shape[2]).astype('uint8')
# Add the Gaussian noise to the image
img_gauss = cv2.add(img,gauss)
#Display the grey scale image
new_img=cv2.cvtColor(img_gauss,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
````
